```python
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
import warnings
warnings.filterwarnings("ignore")

print("Hugging Face setup complete!")
```

```python
generator = pipeline('text-generation', model='gpt2', device=0 if torch.cuda.is_available() else -1)

result = generator(
    "Once upon a time in Ethiopia",
    max_new_tokens=150,
    temperature=0.8,
    top_k=50,
    pad_token_id=50256
)

print(result[0]['generated_text'])
```

```python
# 2. Manual Loading (Better Control)
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

print(f"Model {model_name} loaded successfully!")
print(f"Model parameters: {model.num_parameters():,}")
```

```python
# 3. Text Generation with Manual Control
prompt = "The future of AI in Africa is"
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

outputs = model.generate(
    inputs.input_ids,
    max_new_tokens=200,
    temperature=0.7,
    top_k=50,
    top_p=0.95,
    do_sample=True,
    pad_token_id=tokenizer.eos_token_id
)

generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```
