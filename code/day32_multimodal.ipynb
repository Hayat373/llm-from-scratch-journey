```python
# 1. Image Captioning
captioner = pipeline("image-text-to-text", model="Salesforce/blip-image-captioning-base")

def describe_image(image_url):
    try:
        response = requests.get(image_url, timeout=15)
        response.raise_for_status()
        image = Image.open(BytesIO(response.content))
        
        result = captioner(image, max_new_tokens=50)
        print("✅ Description:", result[0]['generated_text'])
        return result[0]['generated_text']
    except Exception as e:
        print("❌ Error:", e)
        return "Could not load image."

# Reliable Test Images
print("Testing with reliable images...")

describe_image("https://picsum.photos/id/1015/800/600")   # Architecture
ask_about_image("https://picsum.photos/id/1015/800/600", "What is this building?")

describe_image("https://picsum.photos/id/1074/800/600")   # Vehicle
ask_about_image("https://picsum.photos/id/1074/800/600", "What is this vehicle?")

describe_image("https://picsum.photos/id/292/800/600")    # Coffee
ask_about_image("https://picsum.photos/id/292/800/600", "What is this?")
```

```python
# 2. Visual Question Answering with Bus Image

vqa = pipeline("image-text-to-text", model="Salesforce/blip-vqa-base")

def ask_about_image(image_url_or_path, question):
    try:
        if image_url_or_path.startswith("http"):
            response = requests.get(image_url_or_path, timeout=10)
            image = Image.open(BytesIO(response.content))
        else:
            image = Image.open(image_url_or_path)
       
        result = vqa(image, question, max_new_tokens=50)
        print("Question:", question)
        print("Answer:", result[0]['generated_text'])
        return result[0]['generated_text']
    except Exception as e:
        print("Error:", e)
        return "Could not process image."

# Test with Bus Image
bus_image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/0/0f/Bus_in_Addis_Ababa.jpg/800px-Bus_in_Addis_Ababa.jpg"

ask_about_image(bus_image_url, "What is this vehicle?")
ask_about_image(bus_image_url, "What color is the bus?")
ask_about_image(bus_image_url, "Where is this bus located?")
```
