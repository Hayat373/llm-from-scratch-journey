```python
# Day 24 - LLM Evaluation & Benchmarking

import torch
from transformers import pipeline
from sentence_transformers import SentenceTransformer
import numpy as np

print("Evaluation tools loaded!")
```

```python
# Proper Perplexity Calculation
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

def calculate_perplexity(text, model_name="gpt2"):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)
    model = model.to("cuda" if torch.cuda.is_available() else "cpu")

    encodings = tokenizer(text, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model(**encodings, labels=encodings.input_ids)
        loss = outputs.loss

    perplexity = torch.exp(loss).item()
    print(f"Text: '{text[:100]}...'")
    print(f"Perplexity: {perplexity:.2f} (Lower is better)")
    return perplexity

# Test
calculate_perplexity("Ethiopia is a beautiful country with rich history and diverse culture.")
calculate_perplexity("The capital of Ethiopia is Addis Ababa and it is one of the fastest growing cities in Africa.")
```

```python
# 2. Faithfulness & Relevance Evaluation (RAG specific)
def evaluate_rag_answer(question, context, answer):
    # Simple cosine similarity based evaluation
    embedder = SentenceTransformer('all-MiniLM-L6-v2')

    context_emb = embedder.encode(context)
    answer_emb = embedder.encode(answer)

    similarity = np.dot(context_emb, answer_emb) / (np.linalg.norm(context_emb) * np.linalg.norm(answer_emb))

    print(f"Question: {question}")
    print(f"Context similarity score: {similarity:.3f} (higher is better)")
    print(f"Answer: {answer[:200]}...")
    return similarity

# Test
evaluate_rag_answer(
    question="What is the capital of Ethiopia?",
    context="Addis Ababa is the capital and largest city of Ethiopia.",
    answer="The capital of Ethiopia is Addis Ababa."
)
```

```python
# 3. Human-like Evaluation Template
def manual_evaluation(question, answer):
    print("\n=== Manual Evaluation ===")
    print(f"Q: {question}")
    print(f"A: {answer}")
    print("\nRate from 1-10:")
    print("- Relevance:")
    print("- Factual Accuracy:")
    print("- Helpfulness:")
    print("- Naturalness:")

manual_evaluation(
    "Tell me about Ethiopian food",
    "Injera is the national dish of Ethiopia..."
)
```
