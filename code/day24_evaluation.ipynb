```python
# Day 24 - LLM Evaluation & Benchmarking

import torch
from transformers import pipeline
from sentence_transformers import SentenceTransformer
import numpy as np

print("Evaluation tools loaded!")
```

```python
# 1. Perplexity Evaluation (Intrinsic Metric)
def calculate_perplexity(text, model_name="gpt2"):
    generator = pipeline("text-generation", model=model_name, device=-1)
    # Simple approximation
    print(f"Perplexity calculation for: '{text[:100]}...'")
    # In real projects we use proper loss calculation
    return "Perplexity score would be calculated here (lower is better)"

print(calculate_perplexity("Ethiopia is a beautiful country with rich history."))
```

```python
# 2. Faithfulness & Relevance Evaluation (RAG specific)
def evaluate_rag_answer(question, context, answer):
    # Simple cosine similarity based evaluation
    embedder = SentenceTransformer('all-MiniLM-L6-v2')
    
    context_emb = embedder.encode(context)
    answer_emb = embedder.encode(answer)
    
    similarity = np.dot(context_emb, answer_emb) / (np.linalg.norm(context_emb) * np.linalg.norm(answer_emb))
    
    print(f"Question: {question}")
    print(f"Context similarity score: {similarity:.3f} (higher is better)")
    print(f"Answer: {answer[:200]}...")
    return similarity

# Test
evaluate_rag_answer(
    question="What is the capital of Ethiopia?",
    context="Addis Ababa is the capital and largest city of Ethiopia.",
    answer="The capital of Ethiopia is Addis Ababa."
)
```
